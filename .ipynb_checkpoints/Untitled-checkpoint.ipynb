```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df = df.set_index(df.Id, drop=True)
df = df.drop('Id', axis=1)
```

```python
nan_cols = df.isna().any()
non_nan_cols = nan_cols[~nan_cols].index
```

```python
df1 = df[non_nan_cols]
```

# Object columns

```python
obj_col = df1.dtypes == object
obj_col = obj_col[obj_col].index
```

```python
obj_dict = {}
for c in obj_col:
    obj_cnt = df1[c].value_counts()
    plt.title(c)
    obj_cnt.plot.bar()
    plt.show()
```

# Numerical columns

```python
num_col = df1.dtypes != object
num_col = num_col[num_col].index
```

```python
df1[num_col].corr()['SalePrice'].sort_values(ascending=False)
```

```python
df2 = df1[num_col]
```

```python
import seaborn as sns
sns.heatmap(df2.corr())
```

# Modelling

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(df2.drop('SalePrice', axis=1),
                                                  df2.SalePrice,
                                                  test_size=.3,
                                                  random_state=42)
```

# Feature Selection

```python
from sklearn.linear_model import Lasso, Ridge, ElasticNet, LinearRegression
fs_dict = {
    'Lasso' : Lasso(max_iter=10000, alpha=2),
    'Ridge' : Ridge(max_iter=10000),
    'LR' : LinearRegression()
}
for k in fs_dict.keys():
    fs_dict[k].fit(X_train, y_train)
```

```python
tra_col = X_train.columns
coef = pd.Series(fs_dict['Lasso'].coef_).sort_values(ascending=False)
coef.index = tra_col
```

```python
imp_feat = coef[abs(coef) > 1000].index
```

# Modelling
